```python
from sklearn import datasets
from sklearn.model_selection import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16,10)
lr = linear_model.LinearRegression()

y=fss['size_mb'].values
# cross_val_predict returns an array of the same size as `y` where each entry
# is a prediction obtained by cross validation:
x= fss['Columns'].values
predicted = cross_val_predict(lr, fss['Columns'].values.reshape(-1, 1), fss['size_mb'].values, cv=2)

fig, ax = plt.subplots()
ax.scatter(x, y, edgecolors=(0, 0, 0))
ax.plot(x, predicted, 'k--', lw=4)
ax.set_xlabel('#Columns')
ax.set_ylabel('File Size (MB)')
plt.grid()
plt.show()
```

```python
from sklearn import datasets
from sklearn.model_selection import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16,10)
lr = linear_model.LinearRegression()

y=fss['size_mb'].values
# cross_val_predict returns an array of the same size as `y` where each entry
# is a prediction obtained by cross validation:
x= fss['Columns'].values

lr.fit(fss['Columns'].values.reshape(-1, 1),  fss['size_mb'].values)
#predicted = cross_val_predict(lr, fss['Columns'].values.reshape(-1, 1), fss['size_mb'].values, cv=2)
predicted = lr.predict(fss['Columns'].values.reshape(-1, 1))

fig, ax = plt.subplots()
ax.scatter(x, y, edgecolors=(0, 0, 0))
ax.plot(x, predicted, '--', lw=3)
ax.set_xlabel('#Counters',fontsize = 'xx-large')
ax.set_ylabel('File Size (MB)',fontsize = 'xx-large')
plt.title("LTE - Measurement of Cell Performance: File Size Regression ATAE",fontsize = 'xx-large')
plt.grid()
plt.show()
```
